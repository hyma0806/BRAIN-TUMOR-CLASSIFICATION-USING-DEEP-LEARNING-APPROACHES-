# ðŸ“¦ 1. Imports and Setup

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

# 2. Set Image Parameters and Directories

```python
img_height, img_width = 150, 150
batch_size = 32

train_dir = r"E:\brain_tumor_cropped_dataset_split\train"
val_dir = r"E:\brain_tumor_cropped_dataset_split\val"
```

# 3.EDA

# Class Distribution (Image Count per Class)

```python
import os
import matplotlib.pyplot as plt

train_path = r"E:\brain_tumor_cropped_dataset_split\train"
val_path = r"E:\brain_tumor_cropped_dataset_split\val"

def count_images(folder):
    count_dict = {}
    for label in os.listdir(folder):
        label_path = os.path.join(folder, label)
        if os.path.isdir(label_path):
            count = len(os.listdir(label_path))
            count_dict[label] = count
    return count_dict

train_counts = count_images(train_path)
val_counts = count_images(val_path)

# Combine and plot
plt.figure(figsize=(10,5))
labels = list(train_counts.keys())
x = range(len(labels))
plt.bar(x, [train_counts[l] for l in labels], width=0.4, label='Train', align='center')
plt.bar([i + 0.4 for i in x], [val_counts[l] for l in labels], width=0.4, label='Validation', align='center')
plt.xticks([i + 0.2 for i in x], labels)
plt.ylabel("Number of Images")
plt.title("Class Distribution in Train vs Validation")
plt.legend()
plt.show()
```

# Sample Images from Each Class

```python
import matplotlib.image as mpimg
import random

def plot_sample_images(base_dir):
    class_names = os.listdir(base_dir)
    plt.figure(figsize=(12, 8))
    for i, label in enumerate(class_names):
        image_folder = os.path.join(base_dir, label)
        image_file = random.choice(os.listdir(image_folder))
        image_path = os.path.join(image_folder, image_file)
        img = mpimg.imread(image_path)
        plt.subplot(2, 2, i + 1)
        plt.imshow(img)
        plt.title(label)
        plt.axis('off')
    plt.suptitle("Sample Images from Each Class")
    plt.tight_layout()
    plt.show()

plot_sample_images(train_path)
```

# Image Size Distribution

```python
from PIL import Image

img_shapes = []

for cls in os.listdir(train_path):
    cls_path = os.path.join(train_path, cls)
    for img_name in os.listdir(cls_path)[:50]:  # Sample 50 per class for speed
        img_path = os.path.join(cls_path, img_name)
        img = Image.open(img_path)
        img_shapes.append(img.size)

widths, heights = zip(*img_shapes)
plt.figure(figsize=(10,5))
plt.hist(widths, bins=20, alpha=0.7, label='Widths')
plt.hist(heights, bins=20, alpha=0.7, label='Heights')
plt.xlabel("Pixels")
plt.ylabel("Frequency")
plt.title("Image Width and Height Distribution")
plt.legend()
plt.grid(True)
plt.show()
```

# Pixel Intensity Distribution (Grayscale Approximation)

```python
import cv2
import numpy as np

def compute_pixel_histogram(folder):
    pixel_values = []
    for cls in os.listdir(folder):
        cls_path = os.path.join(folder, cls)
        for img_name in os.listdir(cls_path)[:50]:
            img_path = os.path.join(cls_path, img_name)
            img = cv2.imread(img_path)
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            pixel_values.extend(gray.flatten())
    return pixel_values

pixels = compute_pixel_histogram(train_path)

plt.figure(figsize=(8, 4))
plt.hist(pixels, bins=50, color='gray')
plt.title("Grayscale Pixel Intensity Distribution (Train Set)")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

# Aspect Ratio Analysis

```python
aspect_ratios = [w/h for w, h in img_shapes]
plt.figure(figsize=(8, 4))
plt.hist(aspect_ratios, bins=30, color='teal')
plt.title("Aspect Ratio Distribution (Width / Height)")
plt.xlabel("Aspect Ratio")
plt.ylabel("Number of Images")
plt.grid(True)
plt.show()
```

# 4. Data Augmentation and Data Generators

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical'
)
```

# 5. Build the CNN Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, BatchNormalization

cnn_model = Sequential()

cnn_model.add(Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(img_height, img_width, 3)))
cnn_model.add(BatchNormalization())
cnn_model.add(MaxPooling2D(2, 2))

cnn_model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
cnn_model.add(BatchNormalization())
cnn_model.add(MaxPooling2D(2, 2))

cnn_model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
cnn_model.add(BatchNormalization())
cnn_model.add(MaxPooling2D(2, 2))
cnn_model.add(Dropout(0.3))

cnn_model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
cnn_model.add(BatchNormalization())
cnn_model.add(MaxPooling2D(2, 2))
cnn_model.add(Dropout(0.3))

cnn_model.add(Flatten())
cnn_model.add(Dense(128, activation='relu'))
cnn_model.add(Dropout(0.4))
cnn_model.add(Dense(64, activation='relu'))
cnn_model.add(Dense(4, activation='softmax'))  # 4 classes

cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
cnn_model.summary()
```

```python
history_cnn = cnn_model.fit(
    train_generator,
    epochs=30,
    validation_data=val_generator,
    callbacks=[early_stop, reduce_lr],  # same as VGG if defined
    verbose=1
)
```

```python
train_generator_eval = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False  # Disable shuffling here
)
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Predict on training set
train_generator_eval.reset()
Y_train_true = train_generator_eval.classes

Y_train_probs = cnn_model.predict(train_generator_eval, steps=train_generator_eval.samples // batch_size + 1)
Y_train_pred = np.argmax(Y_train_probs, axis=1)

class_labels = list(train_generator_eval.class_indices.keys())

# Classification report
print("ðŸ“‹ CNN Classification Report (Training Set):\n")
print(classification_report(Y_train_true, Y_train_pred, target_names=class_labels))

# Confusion Matrix
conf_matrix_train = confusion_matrix(Y_train_true, Y_train_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_train, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title("CNN Confusion Matrix (Training Set)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot training vs validation accuracy
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history_cnn.history['accuracy'], label='Train Accuracy')
plt.plot(history_cnn.history['val_accuracy'], label='Validation Accuracy')
plt.title('CNN Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.grid(True)

# Plot training vs validation loss
plt.subplot(1, 2, 2)
plt.plot(history_cnn.history['loss'], label='Train Loss')
plt.plot(history_cnn.history['val_loss'], label='Validation Loss')
plt.title('CNN Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.grid(True)

plt.tight_layout()
plt.show()
```

# 6. VGG16 Transfer Learning Code

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten, Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Load base model
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))

# Freeze base model layers
for layer in base_model.layers:
    layer.trainable = False

# Add custom classification head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(4, activation='softmax')(x)

# Final model
vgg_model = Model(inputs=base_model.input, outputs=predictions)

# Compile model
vgg_model.compile(optimizer=Adam(learning_rate=1e-4),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

# Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', patience=3, factor=0.2, verbose=1)

# Train the model
history_vgg = vgg_model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=30,
    validation_data=val_generator,
    validation_steps=len(val_generator),
    callbacks=[early_stop, reduce_lr]
)
```
